```python
def get_labeled_scores(feature_list, importance_dict):
    output_dict = {}
    for i in range(len(feature_list)):
        try:
            output_dict[feature_list[i]] = importance_dict['f' + str(i)]
        except KeyError:
            output_dict[feature_list[i]] = 0
    return output_dict
```

```python
pd.Series(get_labeled_scores(features, bst.get_score(importance_type='gain'))).sort_values()
```

```python
pd.options.display.max_rows = 4000
(pd.Series(get_labeled_scores(features, bst.get_score(importance_type='cover'))) * pd.Series(get_labeled_scores(features, bst.get_score(importance_type='weight'))) * pd.Series(get_labeled_scores(features, bst.get_score(importance_type='gain'))) / 10000000).sort_values()
```

```python
def sample_vecs(df):
    samples = df.loc[:, 'dim0':'dim19'].sample(n=30*df.shape[0], replace=True, weights=df['indprod_num_orders'])
    samples['idx'] = np.arange(0, samples.shape[0]) % df.shape[0]
    samples = samples.groupby('idx').agg('mean').reset_index()
    samples['product_id'] = df.reset_index()['product_id'].astype(int)
    return samples
```
