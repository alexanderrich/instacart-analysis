```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

% matplotlib inline
```

```python
import spacy

nlp = spacy.load('en_vectors_glove_md')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
aisles_df = pd.read_csv("data/aisles.csv")
departments_df = pd.read_csv("data/departments.csv")
products_df = pd.read_csv("data/products.csv")
orders_df = pd.read_csv("data/orders.csv")
prior_df = pd.read_csv("data/order_products__prior.csv")
train_df = pd.read_csv("data/order_products__train.csv")
products_df_merged = (products_df
                      .merge(departments_df, on="department_id")
                      .merge(aisles_df, on="aisle_id"))
```

```python
products_df_merged['allwords'] = products_df_merged.product_name.str.cat([products_df_merged.department, products_df_merged.aisle], sep=' ')
```

```python
products_df_merged.aisle_id.max()
```

```python
vectors = np.array(products_df_merged.allwords.apply(lambda x: nlp(x).vector).tolist())
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=30)
```

```python
pca.fit(vectors)
```

```python
short_vectors = pca.transform(vectors)
```

```python
short_vectors_df = pd.DataFrame(short_vectors)
short_vectors_df.columns = ["dim"+str(c) for c in short_vectors_df.columns]
short_vectors_df['product_id'] = products_df_merged.product_id
```

```python
prior_df = prior_df.merge(orders_df, on='order_id')
```

```python
prior_df['product_id'].name
```

```python
np.log(prior_df.product_id.value_counts()).hist()
```

```python
prior_df = prior_df.merge(products_df_merged[['product_id', 'aisle_id']], on='product_id')
```

```python
prior_day_idx = prior_df[['aisle_id']].join(pd.get_dummies(prior_df.order_dow))
```

```python
aisleprops = prior_day_idx.groupby('aisle_id').agg(np.mean)
```

```python
aisleprops = aisleprops.reset_index()
```

```python
aisleprops = aisleprops.merge(aisles_df, on='aisle_id')
```

```python
aisleprops = aisleprops.drop('aisle', axis=1)
```

```python
aislemelted = aisleprops.melt(id_vars='aisle_id', var_name="order_dow", value_name="aisle_day_prop")
```

```python
prior_df = prior_df.merge(aislemelted, on=['aisle_id', 'order_dow'])
```

```python
np.log(prior_df.aisle_day_prop).mean()
```

```python
prior_vecs = prior_df[['product_id', 'order_dow']].merge(short_vectors_df, on='product_id')
```

```python
del prior_vecs
```

```python
prior_vecs['validation'] = (1.0 * (np.random.random((prior_vecs.shape[0])) > .9))
```

```python
import xgboost as xgb
```

```python
prior_short = prior_vecs.sample(frac=.1).drop(['dim'+str(x) for x in range(5, 30)], axis=1)
```

```python
prior_short.shape
```

```python
d_train = xgb.DMatrix(prior_short.query('validation==0').drop(["product_id", "order_dow", "validation"], axis=1).values,
                          label=prior_short.query('validation==0').order_dow.values)
d_valid = xgb.DMatrix(prior_short.query('validation==1').drop(["product_id", "order_dow", "validation"], axis=1).values,
                          label=prior_short.query('validation==1').order_dow.values)
     
```

```python
prior_short.query('validation==0').order_dow.values.shape
```

```python
prior_short.query('validation==0').drop(["product_id", "order_dow"], axis=1).columns
```

```python

params = {}
params['objective'] = 'multi:softprob'
params['eval_metric'] = 'mlogloss'
params['num_class'] = 7
params['eta'] = 0.05
params['max_depth'] = 3
params['nthread'] = 12
#params['subsample'] = .2
#params['colsample_bytree'] = .5

```

```python
watchlist = [(d_train, 'train'), (d_valid, 'valid')]
```

```python
bst = xgb.train(params, d_train, 1000, watchlist, early_stopping_rounds=50, verbose_eval=2)
```
